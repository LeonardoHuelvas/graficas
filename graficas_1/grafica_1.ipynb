```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración del generador de números aleatorios
rng = np.random.default_rng(42)

# Datos sintéticos de calificaciones de los estudiantes
matematicas = rng.integers(50, 100, 20)
ciencias = rng.integers(40, 95, 20)
literatura = rng.integers(60, 100, 20)

# Datos de errores individuales para el gráfico de barras de error
errores_matematicas = rng.uniform(2, 5, 20)
errores_ciencias = rng.uniform(1, 4, 20)
errores_literatura = rng.uniform(3, 6, 20)

# 1. Gráfico de dispersión --
def crear_grafico_dispersion():
    plt.figure(figsize=(8, 5))
    plt.scatter(matematicas, ciencias, alpha=0.7)
    plt.title("Relación entre las Calificaciones de Matemáticas y Ciencias")
    plt.xlabel("Calificaciones de Matemáticas")
    plt.ylabel("Calificaciones de Ciencias")
    plt.grid(True)
    plt.show()

# Llamamos a las función para generar los gráficos
crear_grafico_dispersion()
```

```python

# 2. Gráfico de barras de error
def crear_grafico_barras_error():
    materias = ['Matemáticas', 'Ciencias', 'Literatura']
    promedios = [np.mean(matematicas), np.mean(ciencias), np.mean(literatura)]
    errores = [np.std(matematicas), np.std(ciencias), np.std(literatura)]

    plt.figure(figsize=(8, 5))
    bar_locations = np.arange(len(materias))
    plt.bar(bar_locations, promedios, yerr=errores, align='center', alpha=0.7, capsize=10)
    plt.xticks(bar_locations, materias)
    plt.title("Calificaciones Promedio y Error Estándar por Materia")
    plt.xlabel("Materias")
    plt.ylabel("Calificaciones Promedio")
    plt.legend(materias)
    plt.show()

# Llamamos a las función para generar los gráficos
crear_grafico_barras_error()
```

```python
# 3. Histograma
def crear_histograma():
    plt.figure(figsize=(8, 5))
    plt.hist(matematicas, bins=8, alpha=0.7)
    plt.title("Distribución de las Calificaciones de Matemáticas")
    plt.xlabel("Calificaciones")
    plt.ylabel("Número de Estudiantes")
    plt.show()

# Llamamos a las función para generar los gráficos
crear_histograma()
```
